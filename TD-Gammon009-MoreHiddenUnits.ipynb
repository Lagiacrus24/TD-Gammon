# Mehr Hidden Units?

In den ersten Versionen von TD-Gammon hat Gerald Tesauro 40 Hidden Units genutzt, in späteren 80 Units.

Bringt die Verdoppelung der Hidden Units etwas?

Um das herauszufinden, erweitere ich die NeuralNetModel Klasse in der init Methode um die Größe der Layer und damit die Daten an verschiedenen Orten gespeichert werden um einen optionalen Namen:

```python
def __init__(self, sess, input_size = 198, hidden_size = 40, output_size = 1, name = None, restore=False)
```

Der Name fällt auf __self.get_name()__ zurück, falls er _None_ ist.

````python
import tensorflow as tf
from NeuralNetModel import TDGammonModel

graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = TDGammonModel(sess, hidden_size = 80, name= "TD-Gammon80", restore=False)
    model.train(20000,1000)
````

### Unterschiede zu 40 Hidden Units

- Nach 5000 Spielen "nur" 85% Siegesrate, nicht 100% 


#### TD-Gammon40 vs TD-Gammon80

````python
import Player
from NeuralNetModel import TDGammonModel

graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = TDGammonModel(sess, restore=True)
    graph2 = tf.Graph()
    sess2 = tf.Session(graph=graph2)
    with sess2.as_default(), graph2.as_default():
        model2 = TDGammonModel(sess2, hidden_size = 80, name = "TD-Gammon80", restore=True)
        model.test(games = 1000, enemyPlayer = Player.ModelPlayer('white', model2))
````

33% ?! Die 40 zusätzlichen Hidden Units machen sich bezahlt.

#### TD-Gammon80 vs Singleton

````python
graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = TDGammonModel(sess, hidden_size = 80, name = "TD-Gammon80", restore=True)
    model.test(games = 1000, enemyPlayer = Player.ValuePlayer('white', Player.singleton))
````

84.7%, ein wenig mehr als mit 40 Hidden Units!

## 2-ply

Vielleicht bringt 2-ply hier ja eine Verbesserung

````python
from NeuralNetModel import TDGammonModel
import tensorflow as tf
import Player

graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = TDGammonModel(sess, hidden_size = 80, name = "TD-Gammon80", restore=True)
    model.test(games = 100, enemyPlayer = Player.TwoPlyModelPlayer('white', model))
````

- Start: 13:36
- Ende: 16:16
- Ergebnis: Leichte Verbesserung

Im Gegensatz zu den 40 Hidden Units ist die Verbesserung recht klein, was aber bedeutet, dass das Netz auch ohne die Gegnerzüge miteinzubeziehen bereits sehr gute Entscheidungen trifft.
