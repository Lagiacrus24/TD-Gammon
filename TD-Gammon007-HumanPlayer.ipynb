# Der menschliche Spieler

Untereinander können die verschiedenen KI beliebig spielen, aber wie gut sie wirklich sind kann man erst feststellen, wenn Menschen gegen die KI spielen

## HumanPlayer

__HumanPlayer__ ist eine Subklasse von Player. Der Spieler erhält eine sortierte, durchnummerierte Liste aller möglichen Züge aus denen dann mit _input_ die Eingabe vom Spieler aus der Kommandozeile gelesen wird.

```python
from Player import Player

class HumanPlayer(Player):
    
    def get_action(self, actions, game):
        # actions sortieren
        actions = sorted(actions)
        # Nummeriert ausgeben
        for i in range(len(actions)):
            print(i, ":", actions[i])
        # Spieler fragen
        index = int(input("Bitte gewünschte Zugnummer angeben:"))
        # Zug aussuchen
        return actions[index]
        
    def get_name(self):
        return "HumanPlayer"
```

```python
import Player
from FasterBackgammon import Game

players = [HumanPlayer(Game.PLAYERS[0]), Player.ValuePlayer(Game.PLAYERS[1], Player.singleton)]
g = Game()
g.play(players, debug=True)
```
