# Das Große Turnier

Hier tritt jeder Player gegen jeden Anderen an um die Performance gegenüber jeglichen Gegnern zu messen und um herauszfinden, wer der beste Player ist.

```python
import Player
import PlayerTest
import tensorflow as tf
import NeuralNetModel
```

## Herausforderer 1: WayToGo

Der erste einfache Player

### Gegner 1: Random

```python
players = [Player.ValuePlayer('black', Player.way_to_go), Player.RandomPlayer('white')]

PlayerTest.test(players, games=1000)
```

### Gegner 2: Singleton

```python
players = [Player.ValuePlayer('black', Player.way_to_go), Player.ValuePlayer('white', Player.singleton)]

PlayerTest.test(players, games=1000)
```

### Gegner 3: Blocker

```python
players = [Player.ValuePlayer('black', Player.way_to_go), Player.ValuePlayer('white', Player.blocker)]

PlayerTest.test(players, games=1000)
```

### Gegner 4: SingleToGo

```python
players = [Player.ValuePlayer('black', Player.way_to_go), Player.ValuePlayer('white', Player.single_to_go)]

PlayerTest.test(players, games=1000)
```

### Gegner 5: TD-Gammon

```python
graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = NeuralNetModel.TDGammonModel(sess, restore=True)
    players = [Player.ValuePlayer('black', Player.way_to_go), Player.ModelPlayer('white', model)]
    PlayerTest.test(players, games=1000)
```

### Gegner 6: TDGammon80

```python
graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = NeuralNetModel.TDGammonModel(sess, hidden_size = 80, name = "TD-Gammon80", restore=True)
    players = [Player.ValuePlayer('black', Player.way_to_go), Player.ModelPlayer('white', model)]
    PlayerTest.test(players, games=1000)
```

### Gegner 7: TF-Gammon

```python
graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = NeuralNetModel.TFGammonModel(sess, restore=True)
    players = [Player.ValuePlayer('black', Player.way_to_go), Player.ModelPlayer('white', model)]
    PlayerTest.test(players, games=1000)
```

### Gegner 8: TF-Gammon2

```python
graph = tf.Graph()
sess = tf.Session(graph=graph)
with sess.as_default(), graph.as_default():
    model = NeuralNetModel.TFGammonModel2(sess, hidden_size = 80, name = "TF-Gammon80", restore=True)
    players = [Player.ValuePlayer('black', Player.way_to_go), Player.ModelPlayer('white', model)]
    PlayerTest.test(players, games=1000)
```
